```r
#Der Alkoholkonsum von den 4 Kategorien(0-39, 40,79, 80,119, 120+) ist ungefähr gleich aufgeteilt innerhalb der jeweiligen Altersgruppen
#Lediglich gibt es einen Ausreiser bei der Altergruppe 75+, hier Konsumieren ein Großteil zwischen 40-79g/Tag.
plot(esoph$agegp, esoph$alcgp, xlab = "Alter", ylab = "Alkoholkonsum")
```

```r
#Die Kombinationen mit 40-79 und 80-119g/Tag Alkoholkonsum haben ein höheres Risiko im Vergleich zu 0-39 und 120+
boxplot(ncases ~ alcgp, data = esoph, main= "Text", xlab = "Alkoholkonsum", ylab = "Fälle von Krebs")
```

```r
#Das Krebsrisiko sinkt mit erhöhtem Tabakkonsum 
boxplot(ncases ~ tobgp, data = esoph, main= "Text", xlab = "Tabakonsum", ylab = "Fälle von Krebs")
```
